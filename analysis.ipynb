```python
#categorical continuous bivariate analysis(t-test) LoanAmount by married and unmarried
from scipy.stats import ttest_ind
```

```python
couples=data[data['Married']=='Yes']
```

```python
singles=data[data['Married']=='No']
```

```python
ttest_ind(couples['LoanAmount'], singles['LoanAmount'] ) 
```

```python
#since the p value is very much less than 0.05, so the difference between the loan amounts of the married and the unmarried people is statistically different.
```
